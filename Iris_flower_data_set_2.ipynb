```python
import pandas as pd
import matplotlib.pyplot as plt
iris = pd.read_csv("/content/IRIS.csv")

iris.describe().plot(kind= "area", fontsize=14,figsize=(20,12), table = True, colormap = "Accent")
plt.ylabel('Values')
plt.xlabel('Statistics')
plt.title('General Statistics of Iris Dataset')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
iris = pd.read_csv("/content/IRIS.csv")
ax= plt.subplot()
sns.countplot('species', data=iris)
plt.title("Iris species Count")
plt.show
```

```python
import pandas as pd
import matplotlib.pyplot as plt
iris = pd.read_csv("/content/IRIS.csv")

iris['species'].value_counts().plot.pie(explode=[0.1,0.1,0.1], shadow=True, fontsize=14 ,figsize=(10,8),autopct='%1.1f%%')
plt.title('Iris Species')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
iris = pd.read_csv("/content/IRIS.csv")

fig = iris[iris.species=='Iris-setosa'].plot(kind='scatter', x='sepal_length', y='sepal_width', color= 'blue', label='Setosa')
iris[iris.species=='Iris-versicolor'].plot(kind='scatter', x='sepal_length', y='sepal_width', color= 'orange', label='Versicolor', ax=fig)
iris[iris.species=='Iris-virginica'].plot(kind='scatter', x='sepal_length', y='sepal_width', color= 'green', label='Virginica', ax=fig)


fig.set_xlabel("Sepal Length")
fig.set_ylabel("Sepal Width")
fig.set_title("Sepat Length Vs Width")
fig = plt.gcf()
fig.set_size_inches(12,8)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
iris = pd.read_csv("/content/IRIS.csv")

fig = sns.jointplot(x='sepal_length',y='sepal_width', data=iris, color='red').plot_joint(sns.kdeplot, zorder=0, n_levels=6)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
iris_v2 = pd.read_csv("/content/IRIS.csv")

g= sns.jointplot(x='sepal_length',y='sepal_width', data=iris_v2,  color='m' , kind='kde')
g.plot_joint(plt.scatter, c='w', s=40, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$SepalsLength(Cm)$","$SepalsWidth(Cm)$")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
iris = pd.read_csv("/content/IRIS.csv")

sub = iris[iris['species']=='Iris-setosa']
sns.kdeplot(data=sub[['sepal_length','sepal_width']], shade=True, shade_lowest=False)
plt.title("Iris-setosa")
plt.xlabel("Length")
plt.ylabel("width")
plt.show()
```
